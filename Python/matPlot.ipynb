```python
import matplotlib.pyplot as plt
import pandas as pd
from pandas_datareader import data
```

```python
stocks = data.DataReader(name='MSFT',data_source='stooq',start='2022-01-01', end='2023-01-01')
stocks=stocks.reset_index()
stocks
```

```python
plt.plot(stocks['Date'], stocks['Close'], color='Red', label='Microsoft Corporation')
plt.show()
```

```python
plt.pie(stocks['Close'])
plt.show()
```

```python
plt.scatter(stocks['Date'], stocks['Close'], color='Green')
plt.show()
```
